## Series como diccionario especializado

De esta manera, se puede pensar en una ``Serie`` Pandas un como un diccionario especializado.

Un diccionario es una estructura que mapea llaves arbitrarias a un conjunto de valores arbitrarios, y una serie es una estructura que mapea llaves de un mismo tipo a un conjunto de valores de un mismo tipo. Este tipado (la exigencia de un tipo definido de dato) es importante: la infomración del tipo de datos de una serie Pandas la hace mucho más eficiente que los diccionarios Python para ciertas operaciones.

```python
goles_favor_dict = {'Perú':2,
          'Brasil':5,
          'Colombia':5,
          'Argentina':3,
          'Uruguay': 5,
          'Chile':0}

goles_favor = pd.Series(goles_favor_dict)
goles_favor
```

```python
goles_favor.sort_values(ascending=False)
```

Puedes notar que los índices han sido ordenados. Ese es el comportamiento por defecto de Pandas.

```python
goles_favor["Perú"]
```

A diferencia de los diccionarios, las ``Series`` pueden soportar operaciones de tipo array tal como el <b>slicing:</b>

```python
goles_favor["Brasil": "Colombia"]
```

 # Data Frames

Un <b>DataFrame</b> son estructuras bidimensionales donde las columnas están etiquetadas con su valor y pueden ser de tipos distintos. Una analogía a un Dataframe podria ser una hoja de cálculo Excel o una tabla de una base de datos.

De forma opcional un Dataframe puede ser un índice. Ese índice será el nombre de las filas.

## DataFrame como un arreglo NumPy generalizado

Si una ``Serie`` es el análogo de un arreglo unidimensional con índices flexibles, un ``DataFrame`` es el análogo de un arreglo bidimensional con índices de fila y nombres de columna flexibles.

```python
# Goles en contra
goles_contra_dict = {'Perú': 2,
        'Brasil':1,
        'Colombia':2,
        'Argentina':5,
        'Uruguay':0,
        'Chile': 0}

goles_contra = pd.Series(goles_contra_dict)
goles_contra
```

**Ceruzado de datos (merge)** similar a la que se puede hacer en BBDD con join.

```python
tabla1 = pd.DataFrame({'key1' : ['A','B','C','D'], 'lval':[1,2,3,4]})
tabla2 = pd.DataFrame({'key2': ['C','D','E','F'], 'rval':[5,6,7,8]})
display(tabla1)
display(tabla2)
new1 = pd.merge(tabla1,tabla2,left_on='key1',right_on='key2', how="left")
# si el campo para fusionar se llama igual en ambas tablas, basta con poner on = ""
#how : forma en la que combinamos con una intersección o con una unión,
#sólo con la izquierda o sólo con la tabla de la derecha
#left, right, inner o outher.
display(new1)
new2 = pd.merge(tabla1,tabla2,left_on='key1', right_on='key2', how="inner")
display(new2)
```

## Agrupaciones

Para hacer agrupaciones, podemos usar **groupby**

```python
f = pd.DataFrame({'A' : ['foo', 'bar'] * 12,
                   'B': ['one', 'two', 'three']*8,
                   'val1' : np.random.rand(24),
                   'val2' : np.random.rand(24)}) 
display(f)
```

```python
group = f.groupby(['A', 'B'])
print("--------------")
display(group)
# los grupos no se pueden mostrar directamente,
# group está pensado para ejecutar posteriormente una acción
# por ejemplo sumar
print("-------------suma agrupada-----------")
display(group.sum())
print("---------estadisticos agrupados------------")
display(group.describe())
```

Una agrupación directa:

```python
f.groupby(['A','B']).sum()
```

Ahora aplicamos un filtro sobre una agrupación (similar al having de SQL):

```python
agrup = f.groupby(['A','B']).sum()
agrup[agrup.val1 > 2.0]
```

¿Te acuerdas de las tablas dinámicas de Excel? Pues, DataFrame tiene una funcionalidad parecida, con la función <b>pivot_table</b>

```python
display(f)
```

```python
#pivot_table tiene 3 partes: index, columns and values(similar a excel)
#margins se refiere a los subtotales
f_pivote = f.pivot_table(index="A", columns="B", values="val1", aggfunc="mean", margins=True, margins_name='Subtotal')
f_pivote
```

```python
#podemos hacer una tabla cruzada simple con 2 variables de un dataframe
pd.crosstab(f.A, f.B)
```

```python
f_pivote.stack()   #devuelve una serie con los datos apilados
```

```python
f_pivote.stack().unstack()  #regresa al dataframe pivoteado
```

## PANDAS

Es un paquete contruido sobre NumPY y proporciona una implementación eficiente de un DataFrame. Estos son esencialmente matrices multidimencionales con etiquetas de fila y columna adjuntas, a menudo, con tipos heterogéneos y/o datos faltantes. Además de ofrecer una interfaz de almacenamiento conveniente para datos etiquetados, Pandas implementa una serie de poderosas operaciones de datos familiares para los usuarios de marcos de base de datos y programas de hojas de cálculo.
