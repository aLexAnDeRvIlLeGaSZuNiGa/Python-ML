# Visualización con Seaborn

Seaborn es más que una mera librería de visualización, es una librería para representación estadística, ya que muestra fácilmente la relación que guardan los datos para detectar tendencias y patrones.

Seaborn es una librería para pintar gráficas construidas a partir de matplotlib.

```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
!pip install pydataset
from pydataset import data #contiene datasets incluidos para pruebas 
iris = data('iris')
iris.head()
```

   + <b>Barplot.</b>Muy recomendable para la visualización de datos agregados, como la media

```python
sns.barplot(x='Species', y='Sepal.Length', data=iris, estimator=np.mean)
plt.show()
```

   +  <b>Histograma.</b>Para conocer la distribución de los datos, el histograma es una herramienta básica, sin embargo, esta representación está muy influenciada por el número de bins que se seleccionen y el ancho de cada uno.

```python
sns.displot(iris['Sepal.Length'])
plt.show()
```

 + <b>KDE plots (Kernel Density Estimation plots).</b>Genera mejores representaciones de distribuciones de datos que un histograma al suavizar la forma de los datos, pero no aporta información estadística.

```python
sns.kdeplot(data=iris['Sepal.Length'], shade=True)
plt.show()
```

 + <b>Box plots.</b>Útiles para conocer los rangos de datos, si existen outliers, la media y el rango intercuartil en el que se destribuyen los datos.

```python
sns.boxplot(data=iris, x='Species', y='Sepal.Length')
plt.show()
```

 + <b>Violin plots.</b>La conjunción de KDE + box plots, clave para comparar distribuciones. Un violin representa la distribución, su media, el rango intercuartil y el intervalo de confianza de 95% en el que se distribuyen los mapas.

```python
sns.violinplot(data=iris, x='Species', y='Sepal.Length')
plt.show()
```

 + <b>Catplot().</b>Este método es muy útil cuando en el análisis se representan variables categóricas. Otra alternativa al diagrama de barras.

```python
sns.catplot(x='Species', y='Sepal.Length', kind="swarm", data=iris)
plt.show()
```

 + <b>Scatter plots.</b>Ideal para la representación de correlaciones y cómo una variable afecta a la otra.

```python
sns.scatterplot(x="Sepal.Length", y="Sepal.Width", data=iris)
plt.show()
```

```python
sns.scatterplot(x="Sepal.Length", y="Sepal.Width", hue='Species', data=iris)
plt.show()
```

<b>Distribuciones Bivariantes</b>

 + El equivalente a plt.scatter de matplotlib en seaborn es sns.joinplot().
 + No solo pinta el scatter sino que de manera automática muestra los histogramas de las variables involucradas

```python
sns.jointplot(x=iris['Sepal.Length'],y=iris['Petal.Length'])
plt.show()
```

```python
sns.jointplot(x=iris['Sepal.Length'],y=iris['Petal.Length'],hue=iris['Species'])
plt.show()
```

```python
sns.jointplot(x=iris['Sepal.Width'],y=iris['Petal.Length'], kind='hex',color='purple')
plt.show()
```
