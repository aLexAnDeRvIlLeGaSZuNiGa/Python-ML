<h2>Estructuras de control de flujo condicinales</h2><br><br>
Las estructuras de control de flujo condicionales, se definen mediante el uso de tres palabras claves reervadas, del lenguaje:<br>if(si), elif(sino,si) y else(sino)<br><br><b>Ejemplo 1:</b>Si semaforo esta en verde, cruzar la calle. Sino, esperar

``if semaforo == 'verde':``<br>
    <li>``print('Cruzar la calle')``<br>
``else:`` <br>
    <li>``print('Esperar')``


Aplicando para semaforo rojo

```python
semaforo='rojo'
```

```python
if semaforo == 'verde':
    print('Cruzar la calle')
else:
    print('Esperar')
```

<li><b>Estructura condicional:</b> Operacion ternaria en una fila (para una condicion)

```python
print('Cruzar la calle') if semaforo == 'verde' else print('Esperar')
```

<b>Ejemplo 2:</b> Si gasto hasta S/.100, pago con dinero en efectivo. Si no, si gasto mas de S/.100 pero menos de S/.300, pago <br> con tarjeta de debito.Si no, pago con tarjeta de credito.<br><br>
<b>Pregunta:</b>¿Si gaste S/.210, con que pago?

```python
compra = 210
if compra <= 100:
    print("Pago en efectivo")
elif compra > 100 and compra < 300:
    print("Pago con tardeja de debito")
else:
    print("Pago con tarjeta de credito")
```

<b>Ejemplo 3:</b> Si la compra es mayor a S/.100, obtengo un descuento del 10%. 

```python
total_compra = 500
if total_compra > 100:
    tasa_descuento = 10
    importe_descuento = total_compra * tasa_descuento / 100
    importe_a_pagar = total_compra - importe_descuento
else:
    importe_a_pagar = total_compra
print(importe_a_pagar)
```

<h2>Estructuras de control iterativas</h2>
<h3>Bucle while</h3>

Este bucle, se encarga de ejecutar una misma accion "mientras que" una determinada condicion se cumpla.

```python
#El while es mejor que el for en funciones
anio = 2005
while anio <= 2018:
    print("Informes de Año", str(anio))
    anio += 1 #anio = anio+1
```

```python
#con for
for anio in range(2005,2018):
     print("Informes de Año", str(anio))
```

```python
#Cuidado con los bucles infinitos
anio = 2005
while anio <= 2018:
    print("Informes de Año", str(anio))
    anio +=1
```

```python
valor = -5
while (valor < 5 ):
    print("valor "+str(valor))
    valor+=1 
#end while
print("Fin del while")
```

```python
# cambiando la tabulacion de una linea, todo cambia
valor = -5
while ( valor < 5):
    print("valor "+str(valor))
    valor+=1
    print("Fin del while")
```

<h2>Bucle for sobre un string (letra por letra)</h2>

```python
for letter in "Sunday":
    print(letter)
```

```python
# letra por letra de forma invertida
day = "Sunday"
day_inv = day[::-1] #invierte el orden de las letras de un string
for letter in day_inv:
    print(letter)
```

```python
#Fibonacci en una lista
n = 20
sequence = [0,1] #es una lista de numeros
for i in range(2,n): # Para todo n>=2
    sequence.append(sequence[i-1]+sequence[i-2])
print(sequence)
```

```python
#Caso especial de fibonacci de 3 elementos
n=20
sequence = [0,1,1]
for i in range(3,n): #Para todo n>=3
    sequence.append(sequence[i-1]+sequence[i-2]+sequence[i-3])
print(sequence)
```
