#### groupby y otros

```python
autos.groupby('cilindraje').sum()
```

```python
autos.groupby('cilindraje').mean()
```

```python
autos.groupby('cilindraje').median()
```

```python
autos.groupby('cilindraje').min()
```

```python
autos.groupby('cilindraje').max()
```

```python
autos.groupby('cilindraje').var()
```

```python
autos.groupby('cilindraje').std()
```

```python
autos.groupby('cilindraje').count()
```

```python
autos.groupby('cilindraje')['desplazamiento'].std()
```

```python
# tambien usar satck en lugar de unstack
autos.groupby('cilindraje')['desplazamiento'].describe().unstack()
```

```python
autos.groupby('cilindraje').aggregate(['min', np.median, max])
```

```python
# Para exportar la data
autos.groupby('cilindraje').aggregate(['min', np.median, max]).to_csv('Agrupacion_1.csv')
```

```python
# Para exportar la data TRANSPUESTA OSEA EN COLUMNA
autos.groupby('cilindraje').aggregate(['min', np.median, max]).T.to_csv('Agrupacion_3.csv')
```

```python
autos.groupby('cilindraje').aggregate({'caballo_fuerza': 'min', 
                             'aceleracion':'mean'})
```

```python
autos.groupby('cilindraje').transform(lambda x: x - x.mean()) #OJO transform aplica a todas las columnas mejor usar el apply o el map
```
